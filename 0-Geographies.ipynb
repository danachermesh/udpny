```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pylab as pl
import seaborn as sns
sns.set_style('whitegrid')
```

```python
Alltracts = gpd.GeoDataFrame.from_file("basemaps/Basemap/basemap.shp")
Alltracts['GEOID'] = Alltracts['GEOID'].apply(lambda x: '{0:0>11}'.format(x))

print(Alltracts.shape)
print(Alltracts.columns)
Alltracts.head(2)
```

```python
fig = pl.figure(figsize=(10,10))
ax = fig.add_subplot(111)

Alltracts.plot(ax=ax, color='w', edgecolor='k',linewidth=.2);
# Counties.plot(ax=ax, color='#EDFF78', edgecolor='k',linewidth=.2, alpha=0.6);
# Places.plot(ax=ax, color='r', edgecolor='k',linewidth=.8);

ax.axis('off')
```
